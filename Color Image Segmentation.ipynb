```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage import io, measure
```

```python
img = io.imread('marbles2.jpg')
plt.imshow(img)
```

```python
#convert rgb to hsv using opencv

hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
#mask = cv2.inRange(hsv,(100, 90, 90), (120, 255, 255)) #Blue
mask = cv2.inRange(hsv,(0, 0, 180), (180, 70, 255)) #White

plt.imshow(mask)
```

```python
#show blue
blue_marbles = cv2.inRange(hsv,(100, 90, 90), (120, 255, 255))
plt.imshow(blue_marbles)
```
